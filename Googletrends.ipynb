```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import time

Trends = TrendReq(hl='en-US', tz=300)
```

```python
keyword = ["gluten"]
Trends.build_payload(keyword, cat=0, timeframe='today 12-m')
time.sleep(5)
```

```python
data = Trends.interest_over_time()
data = data.sort_values(by="gluten", ascending = False)
data = data.head(10)
print(data)
```

```python
keyword = ["gluten"]
Trends.build_payload(keyword, cat=0, timeframe='today 12-m', geo='US', gprop='')
data = Trends.interest_over_time()
data = data.sort_values(by="gluten", ascending = False)
data = data.head(10)
print(data)
```

```python

data = Trends.interest_by_region(resolution='REGION')
data = data.reset_index().rename(columns={'geoName':'states'})
data = data.sort_values(by='gluten', 
                        ascending = False)
data = data.head(10).reset_index(drop=True)
print(data)
```

```python
import matplotlib.pyplot as plt
# Reset index to bring 'geoName' as a regular column
data = data.reset_index()
# Plot the data
data.plot(x='states', y='gluten', figsize=(10,5), kind="bar",color='red')
# Apply the style and show the plot
plt.style.use('ggplot')
plt.show()
```

```python
keyword = ["netflix"]
Trends.build_payload(keyword, cat=0, timeframe='now 7-d', geo='US', gprop='')
#data = Trends.interest_over_time()
data = Trends.interest_by_region(resolution='REGION')
data = data.sort_values(by="netflix", ascending = False)
#data = data.head(10)
data = data.reset_index().rename(columns={'geoName': 'state'})
print(data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetics for the plot
sns.set(style='whitegrid')

# Plot a horizontal bar chart
plt.figure(figsize=(12, 8))  # Set the figure size
sns.barplot(x='netflix', y='state', data=data, palette='viridis')

# Add titles and labels
plt.title('Interest in "Netflix" by State (Past 7 Days)', fontsize=16)
plt.xlabel('Interest Level', fontsize=14)
plt.ylabel('State', fontsize=14)

# Show the plot
plt.show()
```

```python


# Filter the DataFrame to get the least states
top_5 = data.tail(5)

plt.figure(figsize=(10, 6))
sns.barplot(x='netflix', y='state', data=top_5, palette='viridis')

plt.title('Least 5 States', fontsize=16)
plt.xlabel('Interest Level', fontsize=14)
plt.ylabel('State', fontsize=14)

# Show the plot
plt.show()
```
